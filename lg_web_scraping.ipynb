```python

# import BeautifulSoup as soup
from bs4 import BeautifulSoup as soup
from requests import get
import pandas as pd
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from openpyxl import load_workbook
path = r"C:/TV.xlsx"
book = load_workbook(path)
```

```python

tvs_url='https://www.lg.com/in/tvs'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(tvs_url)
time.sleep(5)
driver.find_element_by_link_text("View All").click()
time.sleep(5)
tvs_soup = soup(driver.page_source, 'html.parser')
tvs_containers = tvs_soup.find_all('p', class_ = 'model-name')
print(len(tvs_containers))
head=True
cellRow,cellCol=0,0
```

```python
tvs_df=pd.DataFrame()
for j in range(0,len(tvs_containers)):
    tv_link=tvs_containers[j].a['href']
    tv_url = 'https://www.lg.com'+str(tv_link)
    tv_response = get(tv_url)
    tv_soup = soup(tv_response.text, 'html.parser')
    type(tv_soup)
    tv_size_containers=tv_soup.find_all('a', attrs={'data-link-area':"product_summary-sibling-carrier"})
    
    for z in range(0,len(tv_size_containers)):
        tv_size_link=tv_size_containers[z]['href']
        tv_size_url = 'https://www.lg.com'+str(tv_size_link)
        print(tv_size_url)
        tv_size_response = get(tv_size_url)
        tv_size_soup = soup(tv_size_response.text, 'html.parser')
        head_containers = tv_size_soup.find_all('dt')
        value_containers=tv_size_soup.find_all('dd')
        model_name= tv_size_soup.find_all('div', class_ = 'model-name')
        model_title= tv_size_soup.find_all('div', class_ = 'model-title')
        tv_dict=dict()
        tv_dict['Model name']=[model_name[2].text]
        tv_dict['Model title']=[model_title[0].text]
        
        for i in range(0,len(head_containers)):
            feature=head_containers[i].text.strip()
            value=value_containers[i].text.strip()
            tv_dict[feature]=[value]
            
        tv_df=pd.DataFrame(tv_dict)
        tvs_df=tvs_df.append(tv_df,ignore_index = True)
        
print(tvs_df)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book=book
tvs_df.to_excel(writer, sheet_name = 'LG')
writer.save()
```
