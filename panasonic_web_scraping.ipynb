```python

from bs4 import BeautifulSoup as soup
from requests import get
import pandas as pd
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from openpyxl import load_workbook
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC

path = r"C:/TV.xlsx"
book = load_workbook(path)


```

```python
tvs_df=pd.DataFrame()
driver = webdriver.Chrome(executable_path=r"C:\Users\805938\.wdm\drivers\chromedriver\win32\86.0.4240.22\chromedriver.exe")
tvs_url='https://www.panasonic.com/in/search.html?q=Television&searchIn=&_charset_=UTF-8'
driver.get(tvs_url)   
for i in range (0,4):
    tvs_type_soup = soup(driver.page_source, 'html.parser')
    tvs_type_containers = tvs_type_soup.find_all('dt', class_ = 'subheaderlink')
    print(len(tvs_type_containers))
    
    for j in range(0,len(tvs_type_containers)):
        tvs_url=tvs_type_containers[j].a['href']
        tvs_response = get(tvs_url)
        tvs_soup = soup(tvs_response.text, 'html.parser')
        tvs_containers = tvs_soup.find_all('a', class_ = 'linkarea')
        for z in range(0,len(tvs_containers)):
            tv_link=tvs_containers[z]['href']
            tv_url = 'https://www.panasonic.com'+str(tv_link)
            tv_url=tv_url.replace('html','specs.html')
            if (tv_url.find('televisions') != -1):
                tv_response = get(tv_url)
                tv_soup = soup(tv_response.text, 'html.parser')
                tv_dict=dict()
                table_containers = tv_soup.find_all('tr')
                print(tv_url)
                
                for k in range(1,len(table_containers)):
                    try:
                        head_container=table_containers[k].find_all('th',attrs={'scope':"row"})
                        value_containers = table_containers[k].find_all('td',class_="tb-blk")
                        feature=head_container[-1].text.strip()
                        value=value_containers[-1].text.strip()
                        tv_dict[feature]=[value]
                    except Exception:
                        pass
                    
                tv_df=pd.DataFrame(tv_dict)
                tvs_df=tvs_df.append(tv_df,ignore_index = True)
    
    element= driver.find_element_by_css_selector("a.next")
    driver.execute_script("arguments[0].click();", element)
print(tvs_df)
```

```python
print(tvs_df)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book=book
tvs_df.to_excel(writer, sheet_name = 'PANASONIC')
writer.save()
```
