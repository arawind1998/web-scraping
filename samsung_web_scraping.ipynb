```python

from bs4 import BeautifulSoup as soup
from requests import get
import pandas as pd
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
from openpyxl import load_workbook
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC

path = r"C:/TV.xlsx"
book = load_workbook(path)


```

```python

tvs_url='https://www.samsung.com/in/tvs/all-tvs/'

driver = webdriver.Chrome(executable_path=r"C:\Users\805938\.wdm\drivers\chromedriver\win32\86.0.4240.22\chromedriver.exe")
driver.get(tvs_url)       
time.sleep(10)
for i in range(0,6):
    element= driver.find_element_by_css_selector("a.s-ico-down")
    actions = ActionChains(driver)
    actions.move_to_element(element).perform()
    driver.execute_script("arguments[0].click();", element)
    time.sleep(5)
tvs_soup = soup(driver.page_source, 'html.parser')
tvs_containers = tvs_soup.find_all('a', class_ = 'cm-shop-card__product')
print(len(tvs_containers))
print(tvs_containers)
```

```python

driver = webdriver.Chrome(executable_path=r"C:\Users\805938\.wdm\drivers\chromedriver\win32\86.0.4240.22\chromedriver.exe")
actions = ActionChains(driver)

tvs_df=pd.DataFrame()
for j in range(0,len(tvs_containers)):
    tv_link=tvs_containers[j]['href']
    tv_url = 'https://www.samsung.com/'+str(tv_link)
    driver.get(tv_url) 
    tv_soup = soup(driver.page_source, 'html.parser')
    tv_size_container=tv_soup.find_all('input',class_="s-radio-input")
    for i in range(0,len(tv_size_container)):
        try:
            size_element=driver.find_element_by_id(tv_size_container[i]['id'])
            print(tv_size_container[i]['id'])
            driver.execute_script("arguments[0].click();", size_element)
            time.sleep(4)
            element=driver.find_element_by_css_selector("a.s-ico-down")
            driver.execute_script("arguments[0].click();", element)
            tv_dict=dict()
            head_containers = tv_soup.find_all('em',class_="product-specs__highlights-sub-title")
            value_containers = tv_soup.find_all('span',class_="product-specs__highlights-desc")
            for i in range(0,len(head_containers)):
                    feature=head_containers[i].text.strip()
                    value=value_containers[i].text.strip()
                    tv_dict[feature]=[value]
            tv_df=pd.DataFrame(tv_dict)
            tvs_df=tvs_df.append(tv_df,ignore_index = True)
        except Exception:
            pass
print(tvs_df)
    
    
```

```python
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book=book
tvs_df.to_excel(writer, sheet_name = 'SAMSUNG')
writer.save()
```
