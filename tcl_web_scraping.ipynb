```python

from bs4 import BeautifulSoup as soup
from requests import get
import pandas as pd
from openpyxl import load_workbook
path = r"C:/TV.xlsx"
book = load_workbook(path)

```

```python
tvs_url='https://www.tcl.com/in/en/products.allproducts.html'
tvs_response = get(tvs_url)
tvs_soup = soup(tvs_response.text, 'html.parser')
tvs_containers = tvs_soup.find_all('a', class_ = 'pro-detail')
print(len(tvs_containers))
head=True
cellRow,cellCol=0,0
```

```python
tvs_df=pd.DataFrame()
for j in range(0,len(tvs_containers)):
    tv_link=tvs_containers[j]['href']
    tv_url = 'https://www.tcl.com'+str(tv_link)
    print(tv_url)
    tv_response = get(tv_url)
    tv_soup = soup(tv_response.text, 'html.parser')
    type(tv_soup)
    tv_containers = tv_soup.find_all('div', class_ = 'col-md-6 col-xs-6 list')
    print(type(tv_containers))
    print(len(tv_containers))
    tv_dict=dict()
    for i in range(0,len(tv_containers),2):
        
        feature=tv_containers[i].text
        value=tv_containers[i+1].text
        tv_dict[feature]=[value]
        
    tv_df=pd.DataFrame(tv_dict)
    tvs_df=tvs_df.append(tv_df,ignore_index = True)
print(tvs_df)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book=book
tvs_df.to_excel(writer, sheet_name = 'TCL')
writer.save()

```
